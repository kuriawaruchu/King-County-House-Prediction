```python
data.isnull().sum()
```

**OBSERVATIONS:**
 Based on the results, we can see that there are missing values in two columns: "waterfront" and "yr_renovated".

To handle missing values, there are a few options:

Drop rows with missing values: If the number of missing values is relatively small compared to the total dataset, you can choose to remove those rows. However, be cautious not to lose a significant amount of data.

Impute missing values: If the missing values are relatively few and it's appropriate to estimate or fill in those values, you can use imputation techniques. For numerical columns, common methods include filling missing values with the mean, median, or mode. For categorical columns, you can fill missing values with the most frequent category.

Based on the number of missing values in the "waterfront" and "yr_renovated" columns, here are some practical approaches to consider:

Handling the "waterfront" column:

Since there are 2,376 missing values out of 21,597 total rows, dropping these rows might result in a significant loss of data. Therefore, it might be more practical to impute the missing values.
One approach could be to fill the missing values with the most frequent category in the "waterfront" column. You can use the mode value to replace the missing values.
Alternatively, you can consider creating a new category, such as "Unknown" or "Not Specified," to represent the missing values in the "waterfront" column.
Handling the "yr_renovated" column:

With 3,842 missing values out of 21,597 total rows, there is a substantial amount of missing data in this column.
One option is to drop the "yr_renovated" column entirely since it may not be a critical feature for your analysis. However, if you believe it is important, you can explore imputation methods.
For imputation, you can consider filling the missing values with an appropriate value, such as 0, to indicate that no renovation has been done for the respective properties.

```python
#Handling the "waterfront" column:

#Filling missing values with the mode:
mode_value = data['waterfront'].mode()[0]
data['waterfront'].fillna(mode_value, inplace=True)

#Creating a new category for missing values:
data['waterfront'].fillna('Unknown', inplace=True)
```

```python
#Handling the "yr_renovated" column:
#Filling missing values with 0:

data['yr_renovated'].fillna(0, inplace=True)
```

```python
#After executing the codes above, I verified if the missing values have been handled by running data.isnull().sum()
#again.
data.isnull().sum()
```

However, there are still 63 missing values in the "view" column. Let's discuss how we can handle them.

One approach to handle missing values in the "view" column is to fill them with the most common value (mode) since "view" represents the number of times a house has been viewed. You can use the following code to fill the missing values with the mode:

# Modelling

We need to find the features that are most correlated with the target variable (in this case, 'price')before fitting a regression model because it helps in understanding the relationships between the target variable and the predictor variables. By identifying the features that have a strong correlation with the target variable, we focus on those features during the modeling process.

```python
#Using pandas to find the features that are most correlated with price 
data.corr()['price'].sort_values()
```

We see that Grade, Sqft_living, sqft_living15, sqft_above and bathrooms have a high correlation with price. Sqft_living has a high correlation with other predictor variables and hence will not be used in fitting our model even thougth is has a high correlation to price. 

## Fitting the base model.

We begin by regressing price aganist grade since grade has the highest correlation with price. 

```python
#Fitting a linear regression model to predict the 'price' using the feature 'grade'
X = data[['grade']]
y = data['price']

reg_m = LinearRegression()
reg_m.fit(X,y)
reg_m.score(X, y)

print('The predicted values are : ' + str(reg_m.predict(X)))
print('R-sqd for the linear regression model is : ' + str(reg_m.score(X,y)))
```

#### Fitting a regression plot of price aganist grade

```python
sns.regplot(data['grade'], data['price'])
plt.title("Regression Plot")
print("Grade and price are positively correlated.")
```

Grade only accounts 48.14% of the variation in price besides being the most correlated all other variables held constant. 
A positive correlation can be established given then price increases with an increase in grade. 

#### Multivariate linear regression model

```python
# Fit a  mutlivatriate linear regression model,  
# to predict price against other predictor variables that arehighly correlated to price.

reg_cols = ['bedrooms','bathrooms', 'grade','sqft_above', 'sqft_living15','yr_built']

reg_m1 = LinearRegression()
reg_m1.fit(data[reg_cols],y)

print('The predicted values are : ' + str(reg_m1.predict(data[reg_cols])))
print('R-sqd for the linear regression model is : ' + str(reg_m1.score(data[reg_cols],y)))
```

Upon adding other predictor variables that are highly correlated to price, the model is able to account for 61.87% variation of price. This is an incerease from 48.14% Rsquared after regressing price aganist grade. The R squared is however still low. 

```python
# scikit model
outcome = 'price'

# Fitting the actual model
predictors = '+'.join(reg_cols)
formula = outcome + '~' + predictors
model = ols(formula=formula, data=data).fit()
model.summary()
```

## Assessing the model.

The model is not a good one because it has an r-squared of 0.619, which means that only 61.9% of the variation of price is explained by the model. The intercept coefficient is 109.5771, which represents the predicted value of price when all other independent variables are zero. Despite the model being statistically significant with all the p-values being less than 0.05, normality in the data residuals, which is a regression assumption is violated.

Omnibus test: The Omnibus test statistic is 81.840, and the p-value is 0.000. This suggests that there is evidence against the assumption of normality in the residuals. The residuals may not follow a normal distribution.

Durbin-Watson statistic: The Durbin-Watson statistic is 1.949. A value close to 2 suggests no significant autocorrelation in the residuals. In this case, the value of 1.949 indicates a slight positive autocorrelation.

Jarque-Bera (JB) test: The JB test statistic is 111.366, and the p-value is 6.57e-25. A low p-value indicates that the residuals significantly deviate from a normal distribution.

Skewness: The skewness of the residuals is 0.009, which indicates a slight deviation from perfect symmetry.

Kurtosis: The kurtosis of the residuals is 3.351, which suggests a moderate deviation from the normal distribution. Kurtosis measures the thickness of the tails compared to a normal distribution.

Condition Number: The condition number is 8.25e+03. It measures the sensitivity of the model to changes in the input variables. A high condition number suggests multicollinearity or a high degree of correlation between the input variables.
